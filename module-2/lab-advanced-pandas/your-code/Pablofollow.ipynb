```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
...                          "bar", "bar", "bar", "bar"],
...                    "B": ["one", "one", "one", "two", "two",
...                          "one", "one", "two", "two"],
...                    "C": ["small", "large", "large", "small",
...                          "small", "large", "small", "small",
...                          "large"],
...                    "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
...                    "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})

import pandas as pd
df
```

```python
table = pd.pivot_table(df, index=["A"])
```

```python
table = pd.pivot_table(df, index=["A","B"])
table
```

```python
import numpy as np
table = pd.pivot_table(df, index="A", columns="C" , aggfunc=np.sum#np.mean)
table
```

```python
table = pd.pivot_table(df,values=["E"], index=["A","B"], columns="C" , aggfunc=np.sum ,fill_value=-666)
table
```
