```python
from scipy.interpolate import interp1d
#from scipy.interpolate import interpld

x = np.arange(0, 11, 1)
y = np.cos(x)
f = interp1d(x, y)
xnew = np.arange(0,10.25,0.25)
plt.plot(xnew, f(xnew))
plt.scatter(xnew, f(xnew))
plt.scatter(x, y, color="red")
```

```python
f = interp1d(x, y, 'quadratic')
xnew = np.arange(0, 10.25, 0.25)
plt.plot(xnew, f(xnew))
```
